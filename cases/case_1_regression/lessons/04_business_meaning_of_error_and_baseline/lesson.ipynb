## Two naive guesses

"Always expect the historical average" vs. "always expect on time."

```python
from task import predict_mean_baseline, predict_zero_baseline

mean_pred = predict_mean_baseline(df)
zero_pred = predict_zero_baseline(df)
mean_pred.iloc[0], zero_pred.iloc[0]
```

## Scoring both guesses

MAE is in the same units as the target (minutes) — easy to say out loud to TransLine. RMSE punishes big misses harder.

```python
from task import mean_absolute_error, root_mean_squared_error

actual = df["delay_minutes"]
mean_mae = mean_absolute_error(actual, mean_pred)
mean_rmse = root_mean_squared_error(actual, mean_pred)
zero_mae = mean_absolute_error(actual, zero_pred)
zero_rmse = root_mean_squared_error(actual, zero_pred)
print("mean-baseline  MAE:", mean_mae, " RMSE:", mean_rmse)
print("zero-baseline  MAE:", zero_mae, " RMSE:", zero_rmse)
```
