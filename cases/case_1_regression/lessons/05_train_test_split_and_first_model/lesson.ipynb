# Lesson 5 — Train/Test Split and First Regression Model

Read `README.md` (or `README.pl.md`) first. This is the first model that has to earn its keep against Lesson 4's baseline — on data it never saw during training.

```python
from task import load_clean_shipments, split_shipments

df = load_clean_shipments()
train_df, test_df = split_shipments(df)
len(train_df), len(test_df)
```

## Fit on train, never touch test until scoring

Using the four numeric features Lesson 3 flagged as real signal — not `actual_duration_min`, which we already know is a trap.

```python
from task import fit_model, predict_delay

model = fit_model(train_df)
predicted = predict_delay(model, test_df)
predicted[:5]
```

## Score against a *fair* baseline

Lesson 4's baseline used the mean over the whole dataset. That would leak test information — so this baseline uses only the training mean.

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, root_mean_squared_error

actual = test_df["delay_minutes"]
model_mae = mean_absolute_error(actual, predicted)
model_rmse = root_mean_squared_error(actual, predicted)

train_mean_delay = train_df["delay_minutes"].mean()
baseline_pred = pd.Series(train_mean_delay, index=test_df.index)
baseline_mae = mean_absolute_error(actual, baseline_pred)
baseline_rmse = root_mean_squared_error(actual, baseline_pred)

print("model     MAE:", model_mae, " RMSE:", model_rmse)
print("baseline  MAE:", baseline_mae, " RMSE:", baseline_rmse)
```
