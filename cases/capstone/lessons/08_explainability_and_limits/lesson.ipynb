## A note on scaling

Lesson 4's classification path fits `LogisticRegression` on raw features — the model doesn't need scaling to make correct predictions. This lesson scales anyway, because `reason_codes` below compares coefficient magnitudes across features, and raw coefficients on differently-scaled features (`loan_amount` in the thousands vs. `debt_to_income_ratio` as a small decimal) aren't comparable.

```python
import pandas as pd
from task import reason_codes

scaled_test = pd.DataFrame(
    scaler.transform(test_df[FEATURE_COLUMNS]), columns=FEATURE_COLUMNS, index=test_df.index
)
predictions = model.predict(scaled_test)
denied_row = test_df[predictions == 1].iloc[0]
reason_codes(model, scaler, denied_row, FEATURE_COLUMNS)
```
