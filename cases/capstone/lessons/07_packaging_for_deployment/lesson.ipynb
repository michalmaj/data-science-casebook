## Compare to Lesson 6

Look at your Lesson 6 result for the same dataset. Did adding the categorical
column change things? Two different things changed at once here, not one —
this pipeline's numeric branch also scales features before fitting, which
Lessons 4-6 only did for clustering, not for regression or classification.
For a plain `LinearRegression`, scaling doesn't change predictions at all
(the fitted line is the same either way). For a regularized model like
`LogisticRegression`, it can — so if your classification numbers moved, that
might be the scaling, the added category, or both. There's no single correct
answer to expect across all three dataset choices; check what actually
happened for yours.
