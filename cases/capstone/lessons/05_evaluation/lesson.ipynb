```python
if DATASET_NAME == "clinic_wait_times":
    from task import (
        evaluate_regression,
        fit_regression_baseline_and_model,
        impute_missing,
        load_dataset,
        split_dataset,
    )

    reg_df = load_dataset("clinic_wait_times")
    reg_train_df, reg_test_df = split_dataset(reg_df)
    reg_train_df, reg_test_df = impute_missing(reg_train_df, reg_test_df)
    features = ["num_patients_ahead", "staff_on_duty", "hour_of_day", "patient_age"]
    target = "wait_time_minutes"
    baseline, model = fit_regression_baseline_and_model(reg_train_df, target, features)
    result = evaluate_regression(baseline, model, reg_test_df, target, features)

elif DATASET_NAME == "lendwell_loan_default":
    from task import (
        evaluate_classification,
        fit_classification_baseline_and_model,
        impute_missing,
        load_dataset,
        split_dataset,
    )

    clf_df = load_dataset("lendwell_loan_default")
    clf_train_df, clf_test_df = split_dataset(clf_df)
    clf_train_df, clf_test_df = impute_missing(clf_train_df, clf_test_df)
    features = [
        "loan_amount",
        "annual_income",
        "credit_score",
        "debt_to_income_ratio",
        "employment_years",
        "previous_defaults",
    ]
    target = "defaulted"
    baseline, model = fit_classification_baseline_and_model(clf_train_df, target, features)
    result = evaluate_classification(baseline, model, clf_test_df, target, features)

elif DATASET_NAME == "retail_store_segments":
    from task import (
        evaluate_clustering,
        fit_clustering_model,
        impute_missing,
        load_dataset,
        scale_features,
    )

    cluster_df = load_dataset("retail_store_segments")
    cluster_df, _ = impute_missing(cluster_df, cluster_df)
    features = [
        "store_size_sqft",
        "monthly_revenue",
        "foot_traffic",
        "avg_transaction_value",
        "return_rate",
        "inventory_turnover",
    ]
    scaled_df = scale_features(cluster_df, features)
    model = fit_clustering_model(scaled_df, features)
    result = evaluate_clustering(model, scaled_df, features)

else:
    raise ValueError(f"Unknown DATASET_NAME: {DATASET_NAME!r}")

result
```
